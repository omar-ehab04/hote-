```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
data_no_outliers = data.copy()

for col in numerical_cols:
    data_no_outliers = remove_outliers_iqr(data_no_outliers, col)
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=data_no_outliers[col])
    plt.title(f'Box plot of {col} (Outliers Removed)')
    plt.show()
```

### Data Quality Issues

Based on the exploratory data analysis, the following data quality issues were identified:

*   **Missing Values:** The `children`, `country`, `agent`, and `company` columns contain missing values. The `company` column has a particularly high number of missing values.
*   **Outliers:** Several numerical columns, such as `lead_time`, `stays_in_weekend_nights`, `stays_in_week_nights`, `adults`, `children`, `babies`, `adr`, `previous_cancellations`, and `previous_bookings_not_canceled` contain outliers.

#Phase 2 Data cleaning

```python
data['company'].fillna(0, inplace=True)
data['agent'].fillna(0, inplace=True)
```

```python
print("Missing values after filling with 0:")
print(data[['company', 'agent']].isnull().sum())
```

```python
# Calculate the mode of the 'country' column
mode_country = data['country'].mode()[0]

# Fill the missing values in the 'country' column with the calculated mode
data['country'].fillna(mode_country, inplace=True)
```

```python
# Print the number of missing values in the 'country' column after imputation
print("\nMissing values after imputing with mode:")
print(data['country'].isnull().sum())
```

```python
# Calculate the median of the 'children' column
median_children = data['children'].median()

# Fill the missing values in the 'children' column with the calculated median
data['children'].fillna(median_children, inplace=True)
```

```python
# Print the number of missing values in the 'children' column after imputation
print("\nMissing values after imputing with median:")
print(data['children'].isnull().sum())
```

```python
print("Shape before removing duplicates:", data.shape)
```

```python
data.drop_duplicates(inplace=True)
print("Shape after removing duplicates:", data.shape)
```

```python
numerical_cols = data.select_dtypes(include=np.number).columns.tolist()
data_no_outliers_final = data.copy()

for col in numerical_cols:
    data_no_outliers_final = remove_outliers_iqr(data_no_outliers_final, col)

print("Shape after removing outliers:", data_no_outliers_final.shape)
```
